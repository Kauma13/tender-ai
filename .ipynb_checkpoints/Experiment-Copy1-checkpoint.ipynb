```python
@dataclass(frozen=True)
class HistoricalRecord:
    source_id: str
    domain: str
    question: str
    answer: str
```

```python
def load_historical_records(df: pd.DataFrame) -> List[HistoricalRecord]:
    records: List[HistoricalRecord] = []
    for _, row in df.iterrows():
        q = str(row["Question"]).strip()
        a = str(row["Answer"]).strip()
        if not q or not a:
            continue
        records.append(
            HistoricalRecord(
                source_id=str(row["Source_ID"]).strip(),
                domain=str(row["Domain"]).strip(),
                question=q,
                answer=a,
            )
        )
    return records
```

```python
def to_documents(records: List[HistoricalRecord]) -> List[Document]:
    docs: List[Document] = []
    for r in records:
        text = f"QUESTION:\n{r.question}\n\nANSWER:\n{r.answer}"
        docs.append(
            Document(
                page_content=text,
                metadata={
                    "source_id": r.source_id,
                    "domain": r.domain,
                    "historical_question": r.question,
                    "historical_answer": r.answer,
                },
            )
        )
    return docs
```

```python
from pydantic import BaseModel, Field
from typing import Optional, Literal, Dict, Any
```

```python
class QuestionItem(BaseModel):
    question_id: str
    question: str
    suggested_domain: Optional[str] = None
```

```python
class RetrievedMatch(BaseModel):
    source_id: str
    domain: str
    historical_question: str
    historical_answer: str
    similarity: float
```

```python
class QuestionResult(BaseModel):
    question_id: str
    question: str
    domain: str
    matches: list[RetrievedMatch] = Field(default_factory=list)
    flags: list[str] = Field(default_factory=list)
    error: Optional[str] = None
```

```python
class RunSummary(BaseModel):
    total: int = 0
    flagged: int = 0
    errors: int = 0
    status: Literal["COMPLETED", "COMPLETED_WITH_FLAGS", "PARTIAL"] = "COMPLETED"
```

```python
class TenderState(BaseModel):
    request_id: str
    questions: list[QuestionItem] = Field(default_factory=list)
    results: list[QuestionResult] = Field(default_factory=list)
    summary: RunSummary = Field(default_factory=RunSummary)
    retrieval_top_k: int = 3
    retrieval_threshold: float = 0.65
```

```python
def classify_domain_stub(q: str, suggested: Optional[str]) -> str:
    if suggested:
        return suggested
    s = q.lower()
    if any(k in s for k in ["encrypt", "iso", "incident", "pen", "security"]):
        return "Security & Compliance"
    if any(k in s for k in ["architecture", "deployment"]):
        return "Architecture"
    if any(k in s for k in ["pii", "residency", "retention", "data"]):
        return "Data Governance"
    if any(k in s for k in ["implementation", "timeline", "governance"]):
        return "Delivery"
    if any(k in s for k in ["support", "uptime"]):
        return "Operations & Support"
    if any(k in s for k in ["ai", "hallucination"]):
        return "AI & Data"
    if any(k in s for k in ["risk", "assumption"]):
        return "Commercial & Risk"
    if any(k in s for k in ["training", "change", "adoption"]):
        return "Change & Training"
    return "General"
```

```python
def retrieve_node(state: TenderState) -> TenderState:
    out: list[QuestionResult] = []
    for item in state.questions:
        try:
            domain = classify_domain_stub(item.question, item.suggested_domain)
            hits = vectordb.similarity_search_with_relevance_scores(item.question, k=state.retrieval_top_k)

            matches: list[RetrievedMatch] = []
            for doc, score in hits:
                md = doc.metadata or {}
                matches.append(
                    RetrievedMatch(
                        source_id=str(md.get("source_id", "")),
                        domain=str(md.get("domain", "")),
                        historical_question=str(md.get("historical_question", "")),
                        historical_answer=str(md.get("historical_answer", "")),
                        similarity=float(score),
                    )
                )

            flags = []
            # top_score = matches[0].similarity if matches else 0.0
            # if top_score < state.retrieval_threshold:
            #     flags.append("NO_STRONG_HISTORICAL_MATCH")


            top_score = matches[0].similarity if matches else 0.0

            strong_th = 0.70
            weak_th = 0.60
            
            if top_score >= strong_th:
                flags.append("HISTORICAL_MATCH_STRONG")
            elif top_score >= weak_th:
                flags.append("HISTORICAL_MATCH_WEAK")
            else:
                flags.append("NO_STRONG_HISTORICAL_MATCH")

            out.append(
                QuestionResult(
                    question_id=item.question_id,
                    question=item.question,
                    domain=domain,
                    matches=matches,
                    flags=flags,
                )
            )
        except Exception as e:
            out.append(
                QuestionResult(
                    question_id=item.question_id,
                    question=item.question,
                    domain=item.suggested_domain or "Unknown",
                    matches=[],
                    flags=["RETRIEVAL_ERROR"],
                    error=str(e),
                )
            )
    state.results = out
    return state
```

```python
def summarise_node(state: TenderState) -> TenderState:
    state.summary.total = len(state.questions)
    state.summary.errors = sum(1 for r in state.results if r.error)
    state.summary.flagged = sum(1 for r in state.results if r.flags)

    if state.summary.errors > 0:
        state.summary.status = "PARTIAL"
    elif state.summary.flagged > 0:
        state.summary.status = "COMPLETED_WITH_FLAGS"
    else:
        state.summary.status = "COMPLETED"
    return state
```

```python
from langgraph.graph import StateGraph, END
import uuid
```

```python
graph = StateGraph(TenderState)
graph.add_node("retrieve", retrieve_node)
graph.add_node("summarise", summarise_node)
```

```python
graph.set_entry_point("retrieve")
graph.add_edge("retrieve", "summarise")
graph.add_edge("summarise", END)
```

```python
app_graph = graph.compile()
```

## Task 3 — Domain Classifier Agent (Rules + LLM)

```python
DOMAINS = [
    "Security & Compliance",
    "Architecture",
    "Data Governance",
    "Delivery",
    "Operations & Support",
    "AI & Data",
    "Commercial & Risk",
    "Change & Training",
    "General",
]
```

```python
import re
from dataclasses import dataclass
from typing import Optional, List, Dict, Tuple

DOMAIN_KEYWORDS: Dict[str, List[str]] = {
    "Security & Compliance": ["security", "iso", "soc", "encrypt", "encryption", "tls", "aes", "incident", "breach", "vulnerability", "pen", "penetration", "audit", "compliance"],
    "Architecture": ["architecture", "deployment", "saas", "on-prem", "hybrid", "components", "integration", "api", "identity", "sso"],
    "Data Governance": ["pii", "privacy", "data residency", "residency", "retention", "deletion", "offboarding", "data hosting", "subcontractor", "subprocessor"],
    "Delivery": ["implementation", "timeline", "plan", "methodology", "governance model", "roles", "responsibilities", "reporting cadence", "project"],
    "Operations & Support": ["support", "escalation", "response time", "sla", "uptime", "availability", "monitoring", "service desk"],
    "AI & Data": ["ai", "ml", "machine learning", "llm", "rag", "hallucination", "evaluation", "data leakage", "model"],
    "Commercial & Risk": ["assumption", "constraint", "dependency", "risk", "mitigation", "commercial", "pricing"],
    "Change & Training": ["change", "training", "adoption", "comms", "communications", "stakeholder"],
}

RISK_PATTERNS: List[Tuple[str, str]] = [
    ("CERTIFICATION_CLAIM", r"\b(iso\s*27001|soc\s*2|certificat(e|ion)|certified)\b"),
    ("SLA_PROMISE", r"\b(uptime|availability|sla|response time|guarantee)\b"),
    ("LEGAL_COMPLIANCE", r"\b(gdpr|privacy act|regulatory|legislation|compliance)\b"),
]
```

```python
@dataclass
class DomainDecision:
    domain: str
    confidence: float
    rationale: str
    used: str  # "rules" or "llm"
    risk_hints: List[str]

def detect_risks(question: str) -> List[str]:
    q = question.lower()
    hints = []
    for label, pattern in RISK_PATTERNS:
        if re.search(pattern, q, flags=re.IGNORECASE):
            hints.append(label)
    return sorted(set(hints))

def rule_classify(question: str) -> DomainDecision:
    q = question.lower()

    scores: Dict[str, int] = {d: 0 for d in DOMAINS}
    for domain, kws in DOMAIN_KEYWORDS.items():
        for kw in kws:
            if kw in q:
                scores[domain] += 1

    # pick the best non-General
    best_domain = max(scores, key=lambda d: scores[d])
    best_score = scores[best_domain]

    # determine ambiguity: low score or tie
    sorted_scores = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    top2 = sorted_scores[:2]
    tie = len(top2) == 2 and top2[0][1] == top2[1][1] and top2[0][1] > 0

    risk_hints = detect_risks(question)

    if best_score == 0:
        return DomainDecision(
            domain="General",
            confidence=0.40,
            rationale="No strong keyword signal; defaulting to General.",
            used="rules",
            risk_hints=risk_hints,
        )

    if tie or best_score == 1:
        # ambiguous; return a low-confidence rules decision (LLM can refine)
        return DomainDecision(
            domain=best_domain,
            confidence=0.55,
            rationale=f"Keyword signal is weak/ambiguous (top score={best_score}).",
            used="rules",
            risk_hints=risk_hints,
        )

    # strong enough
    conf = min(0.90, 0.60 + 0.10 * best_score)  # simple scaling
    return DomainDecision(
        domain=best_domain,
        confidence=conf,
        rationale=f"Matched {best_score} domain keywords for '{best_domain}'.",
        used="rules",
        risk_hints=risk_hints,
    )
```

```python
test_questions = [
    "Describe your organisation’s information security management approach.",
    "Do you hold ISO 27001 certification? Provide certificate number and expiry date.",
    "Provide an implementation approach and indicative timeline for delivery within 12 weeks.",
    "What uptime commitment do you offer and how is uptime measured and reported?",
    "How do you use AI/ML in your solution? Include safeguards against hallucinations and data leakage.",
]

for q in test_questions:
    d = rule_classify(q)
    print("\nQ:", q)
    print("->", d)
```

```python
from pydantic import BaseModel, Field
from typing import Literal
from langchain_openai import ChatOpenAI

class LLMClassification(BaseModel):
    domain: str = Field(description=f"One of: {', '.join(DOMAINS)}")
    confidence: float = Field(ge=0.0, le=1.0, description="Confidence 0..1")
    rationale: str = Field(description="Short explanation (1-2 lines)")
    risk_hints: List[str] = Field(default_factory=list, description="Zero or more risk hint labels")

llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)
```

```python
def llm_classify(question: str) -> DomainDecision:
    system = (
        "You are classifying tender questionnaire items into a fixed domain taxonomy. "
        "Return only the requested structured fields."
    )

    user = f"""
    Classify the following question into one domain from this list:
    {DOMAINS}
    
    Also detect risk hints from this allowed list:
    [CERTIFICATION_CLAIM, SLA_PROMISE, LEGAL_COMPLIANCE]
    
    Question:
    {question}
    """.strip()
    
    structured = llm.with_structured_output(LLMClassification)
    out: LLMClassification = structured.invoke([("system", system), ("user", user)])
    
    # Hard safety: if model outputs unknown domain, clamp to General
    domain = out.domain if out.domain in DOMAINS else "General"
    risks = [r for r in out.risk_hints if r in {"CERTIFICATION_CLAIM", "SLA_PROMISE", "LEGAL_COMPLIANCE"}]

    return DomainDecision(
        domain=domain,
        confidence=float(out.confidence),
        rationale=out.rationale.strip(),
        used="llm",
        risk_hints=sorted(set(risks)),
    )
```

```python
def hybrid_classify(question: str) -> DomainDecision:
    rule_decision = rule_classify(question)

    # if rules are strong enough, skip LLM
    if rule_decision.confidence >= 0.75 and not (rule_decision.domain == "General"):
        return rule_decision

    # Otherwise ask LLM, but keep rules risks as baseline
    try:
        llm_decision = llm_classify(question)
        # merge risk hints (union)
        merged_risks = sorted(set(rule_decision.risk_hints) | set(llm_decision.risk_hints))
        llm_decision.risk_hints = merged_risks
        return llm_decision
    except Exception as e:
        # fallback gracefully to rules
        fallback = rule_decision
        fallback.rationale += f" (LLM fallback due to error: {e})"
        return fallback
```

## Task 4 — LangGraph flow: classify → retrieve → decide branch (STRONG/WEAK/NONE)

```python
RISK_PATTERNS = [
    ("CERTIFICATION_CLAIM", r"\b(iso\s*27001|soc\s*2|certificat(e|ion)|certified)\b"),
    ("SLA_PROMISE", r"\b(uptime|availability|sla|response time|guarantee)\b"),
    ("LEGAL_COMPLIANCE", r"\b(gdpr|privacy act|regulatory|legislation|pci\s*dss|hipaa)\b"),
]
```

```python
def match_strength(top_score: float, strong: float = 0.70, weak: float = 0.60) -> str:
    if top_score >= strong:
        return "STRONG"
    if top_score >= weak:
        return "WEAK"
    return "NONE"

def route_from_strength(strength: str) -> str:
    return "HISTORICAL_GUIDED" if strength in {"STRONG", "WEAK"} else "TEMPLATE_SAFE"
```

```python
from pydantic import BaseModel, Field
from typing import Optional, List, Literal, Dict, Any

class QuestionItem(BaseModel):
    question_id: str
    question: str
    suggested_domain: Optional[str] = None

class RetrievedMatch(BaseModel):
    source_id: str
    domain: str
    historical_question: str
    historical_answer: str
    similarity: float

# class EnrichedQuestion(BaseModel):
#     question_id: str
#     question: str
#     domain: str
#     domain_confidence: float
#     domain_used: Literal["rules", "llm"]
#     domain_rationale: str
#     risk_hints: List[str] = Field(default_factory=list)

#     matches: List[RetrievedMatch] = Field(default_factory=list)
#     top_similarity: float = 0.0
#     match_strength: Literal["STRONG", "WEAK", "NONE"] = "NONE"
#     route: Literal["HISTORICAL_GUIDED", "TEMPLATE_SAFE"] = "TEMPLATE_SAFE"

#     error: Optional[str] = None

class EnrichedQuestion(BaseModel):
    question_id: str
    question: str

    # classification
    domain: str
    domain_confidence: float
    domain_used: Literal["rules", "llm"]
    domain_rationale: str
    risk_hints: List[str] = Field(default_factory=list)

    # retrieval
    matches: List[RetrievedMatch] = Field(default_factory=list)
    top_similarity: float = 0.0
    match_strength: Literal["STRONG", "WEAK", "NONE"] = "NONE"
    route: Literal["HISTORICAL_GUIDED", "TEMPLATE_SAFE"] = "TEMPLATE_SAFE"

    # NEW: generation outputs
    answer: str = ""
    confidence: float = 0.0
    alignment: Literal["HIGH", "MED", "LOW"] = "LOW"
    flags: List[str] = Field(default_factory=list)

    error: Optional[str] = None

class RunSummary(BaseModel):
    total: int = 0
    errors: int = 0
    routed_historical: int = 0
    routed_template: int = 0
    flagged_risk: int = 0
    status: Literal["COMPLETED", "PARTIAL"] = "COMPLETED"

class TenderState(BaseModel):
    request_id: str
    questions: List[QuestionItem] = Field(default_factory=list)
    enriched: List[EnrichedQuestion] = Field(default_factory=list)
    summary: RunSummary = Field(default_factory=RunSummary)

    retrieval_top_k: int = 3
    strong_threshold: float = 0.70
    weak_threshold: float = 0.60

# class TenderState(BaseModel):
#     request_id: str
#     questions: List[QuestionItem] = Field(default_factory=list)

#     # Store as dicts to avoid nested model validation issues in LangGraph
#     enriched: List[dict] = Field(default_factory=list)

#     summary: RunSummary = Field(default_factory=RunSummary)

#     retrieval_top_k: int = 3
#     strong_threshold: float = 0.70
#     weak_threshold: float = 0.60
```

```python
def classify_retrieve_route_node(state: TenderState) -> TenderState:
    out: List[EnrichedQuestion] = []

    for item in state.questions:
        try:
            # 1) Domain classify (hybrid)
            dec = hybrid_classify(item.question)

            # 2) Retrieve top-k historical matches
            hits = vectordb.similarity_search_with_relevance_scores(
                item.question, k=state.retrieval_top_k
            )

            matches: List[RetrievedMatch] = []
            for doc, score in hits:
                md = doc.metadata or {}
                matches.append(
                    RetrievedMatch(
                        source_id=str(md.get("source_id", "")),
                        domain=str(md.get("domain", "")),
                        historical_question=str(md.get("historical_question", "")),
                        historical_answer=str(md.get("historical_answer", "")),
                        similarity=float(score),
                    )
                )

            top_sim = matches[0].similarity if matches else 0.0
            strength = match_strength(top_sim, strong=state.strong_threshold, weak=state.weak_threshold)
            route = route_from_strength(strength)

            out.append(
                EnrichedQuestion(
                    question_id=item.question_id,
                    question=item.question,
                    domain=dec.domain,
                    domain_confidence=dec.confidence,
                    domain_used=dec.used,  # "rules" or "llm"
                    domain_rationale=dec.rationale,
                    risk_hints=dec.risk_hints,
                    matches=matches,
                    top_similarity=top_sim,
                    match_strength=strength,
                    route=route,
                )
            )

        except Exception as e:
            out.append(
                EnrichedQuestion(
                    question_id=item.question_id,
                    question=item.question,
                    domain=item.suggested_domain or "General",
                    domain_confidence=0.0,
                    domain_used="rules",
                    domain_rationale="Failed during classification/retrieval.",
                    risk_hints=[],
                    matches=[],
                    top_similarity=0.0,
                    match_strength="NONE",
                    route="TEMPLATE_SAFE",
                    error=str(e),
                )
            )

    state.enriched = out
    return state
```

```python
def summarise_node(state: TenderState) -> TenderState:
    state.summary.total = len(state.questions)
    state.summary.errors = sum(1 for x in state.enriched if x.error)

    state.summary.routed_historical = sum(1 for x in state.enriched if x.route == "HISTORICAL_GUIDED")
    state.summary.routed_template = sum(1 for x in state.enriched if x.route == "TEMPLATE_SAFE")
    state.summary.flagged_risk = sum(1 for x in state.enriched if x.risk_hints)

    state.summary.status = "PARTIAL" if state.summary.errors > 0 else "COMPLETED"
    return state
```

```python
from langgraph.graph import StateGraph, END
import uuid

graph = StateGraph(TenderState)
graph.add_node("classify_retrieve_route", classify_retrieve_route_node)
graph.add_node("summarise", summarise_node)

graph.set_entry_point("classify_retrieve_route")
graph.add_edge("classify_retrieve_route", "summarise")
graph.add_edge("summarise", END)

app_graph = graph.compile()

state = TenderState(
    request_id=str(uuid.uuid4()),
    questions=[
        QuestionItem(question_id="Q001", question="Describe your organisation’s information security management approach."),
        QuestionItem(question_id="Q002", question="Do you hold ISO 27001 certification? Provide certificate number and expiry date."),
        QuestionItem(question_id="Q003", question="Confirm that all customer data is encrypted in transit and at rest. Specify TLS and AES standards."),
        QuestionItem(question_id="Q999", question="Explain your strategy for quantum-safe encryption and post-quantum cryptography readiness."),
    ],
    retrieval_top_k=3,
    strong_threshold=0.70,
    weak_threshold=0.60,
)
```

## Task 5 — Generate answers + guardrails + scoring

```python
from pydantic import BaseModel, Field
from typing import Optional, List, Literal

class EnrichedQuestion(BaseModel):
    question_id: str
    question: str

    # classification
    domain: str
    domain_confidence: float
    domain_used: Literal["rules", "llm"]
    domain_rationale: str
    risk_hints: List[str] = Field(default_factory=list)

    # retrieval
    matches: List[RetrievedMatch] = Field(default_factory=list)
    top_similarity: float = 0.0
    match_strength: Literal["STRONG", "WEAK", "NONE"] = "NONE"
    route: Literal["HISTORICAL_GUIDED", "TEMPLATE_SAFE"] = "TEMPLATE_SAFE"

    # NEW: generation outputs
    answer: str = ""
    confidence: float = 0.0
    alignment: Literal["HIGH", "MED", "LOW"] = "LOW"
    flags: List[str] = Field(default_factory=list)

    error: Optional[str] = None
```

```python
DOMAIN_TEMPLATES = {
    "Security & Compliance": (
        "We follow established information security practices including governance, access control, "
        "logging/monitoring, vulnerability management, and incident response. Specific controls and "
        "evidence (e.g., policies, audit reports) can be provided during due diligence, subject to confidentiality."
    ),
    "Architecture": (
        "We propose a modular architecture aligned to the required deployment model (SaaS/on-prem/hybrid). "
        "Key components include identity/access management, integration interfaces (APIs), data storage, and reporting. "
        "A detailed architecture can be provided following discovery and solution design."
    ),
    "Data Governance": (
        "We manage data securely using least-privilege access, auditing, and secure handling processes. "
        "Data residency, retention, deletion, and offboarding are aligned to customer requirements and applicable regulations. "
        "Subprocessors and hosting details can be disclosed during due diligence."
    ),
    "Delivery": (
        "We deliver using a phased approach: kickoff and planning, requirements confirmation, configuration/integration, "
        "testing (SIT/UAT), training, and go-live readiness. Timelines depend on scope stability, stakeholder availability, "
        "and environment readiness."
    ),
    "Operations & Support": (
        "We provide a tiered support model with defined escalation paths. Service levels (response targets, availability) "
        "are agreed based on the selected support plan and are formalised contractually. Monitoring and reporting are provided "
        "through agreed service reporting."
    ),
    "AI & Data": (
        "Where AI is used, it is governed with access controls, logging, and evaluation. Outputs are reviewed by SMEs where required. "
        "Safeguards include curated test sets, prompt/data controls, and measures to reduce hallucinations and prevent sensitive data leakage."
    ),
    "Commercial & Risk": (
        "We document assumptions, constraints, dependencies, and risks early and review them regularly through governance forums. "
        "Mitigations include staged delivery, clear change control, and early integration and data readiness activities."
    ),
    "Change & Training": (
        "We support change through stakeholder engagement, communications planning, role-based training, and readiness checks. "
        "Training materials and sessions are aligned to user roles, with reinforcement and feedback loops around go-live."
    ),
    "General": (
        "We can address this requirement based on discovery and confirmation of scope, constraints, and stakeholder expectations. "
        "Further detail can be provided during due diligence and solution design."
    ),
}
```

```python
def apply_guardrails(question: str, draft_answer: str, risk_hints: List[str]) -> tuple[str, List[str]]:
    """
    Modify answer if it risks fabricating claims.
    Returns (safe_answer, added_flags)
    """
    flags = []

    q = question.lower()
    a = draft_answer

    # Guardrail 1: Certification claim - avoid inventing certificate numbers/dates
    if "CERTIFICATION_CLAIM" in risk_hints:
        # if question asks for certificate number/expiry -> must be cautious
        if re.search(r"(certificate number|expiry|expiration|valid until)", q):
            flags.append("NEEDS_EVIDENCE")
            flags.append("HIGH_RISK")
            a = (
                "We can provide relevant security certifications and supporting evidence as part of the formal due diligence process "
                "and subject to confidentiality. Certification details (e.g., certificate number and expiry) are provided through "
                "official documentation channels upon request."
            )

    # Guardrail 2: SLA promise - avoid hard guarantees unless supported
    if "SLA_PROMISE" in risk_hints:
        # remove/avoid hard guarantee language
        if re.search(r"\b(guarantee|guaranteed|100%|always)\b", a, flags=re.IGNORECASE):
            flags.append("HIGH_RISK")
            a = re.sub(r"\bguarantee(d)?\b", "target", a, flags=re.IGNORECASE)

        # If question explicitly asks for uptime commitment, be conditional
        if "uptime" in q or "availability" in q:
            flags.append("NEEDS_CONTRACTUAL_CONFIRMATION")
            a = (
                "Availability and service levels are agreed based on the selected hosting and support plan and are formalised contractually. "
                "Uptime is measured using agreed monitoring windows and reported through service reporting. Planned maintenance exclusions and "
                "measurement definitions are documented in the service agreement."
            )

    return a, sorted(set(flags))
```

```python
from langchain_openai import ChatOpenAI

gen_llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.2)

def build_historical_guided_prompt(q: EnrichedQuestion) -> str:
    # Use top matches as evidence
    evidence_blocks = []
    for m in q.matches[:3]:
        evidence_blocks.append(
            f"- Source: {m.source_id} | Domain: {m.domain} | Similarity: {m.similarity:.3f}\n"
            f"  Historical Q: {m.historical_question}\n"
            f"  Historical A: {m.historical_answer}\n"
        )
    evidence = "\n".join(evidence_blocks)

    return f"""
    You are generating a tender response. Use the historical answers as guidance and remain consistent with them.
    Do NOT invent certifications, certificate numbers, audit results, uptime SLAs, or legal compliance claims unless explicitly supported by the evidence.
    Keep the response professional, concise, and aligned to the question.
    
    Question:
    {q.question}
    
    Detected Domain: {q.domain}
    Risk Hints: {q.risk_hints}
    
    Historical Evidence:
    {evidence}
    
    Write a response that answers the question directly. If the evidence is insufficient for a claim, state that details can be provided during due diligence or subject to contract.
    """.strip()


def build_template_safe_prompt(q: EnrichedQuestion) -> str:
    template = DOMAIN_TEMPLATES.get(q.domain, DOMAIN_TEMPLATES["General"])
    return f"""
    You are generating a tender response where there is no strong historical match.
    Use the provided safe template as a baseline, adapt it to the question, and keep language cautious.
    Do NOT invent certifications, certificate numbers, audit results, uptime SLAs, or legal compliance claims.
    
    Question:
    {q.question}
    
    Detected Domain: {q.domain}
    Risk Hints: {q.risk_hints}
    
    Safe Baseline Template:
    {template}
    
    Write a professional response. If specifics are required, state they can be provided during due diligence or agreed contractually.
    """.strip()
```

```python
def compute_confidence(q: EnrichedQuestion) -> float:
    base = 0.45

    # domain signal
    base += 0.25 * float(q.domain_confidence)

    # retrieval signal
    if q.match_strength == "STRONG":
        base += 0.25
    elif q.match_strength == "WEAK":
        base += 0.15
    else:
        base += 0.05

    # risk penalties
    if "HIGH_RISK" in q.flags:
        base -= 0.20
    if "NEEDS_EVIDENCE" in q.flags:
        base -= 0.10
    if "NO_STRONG_HISTORICAL_MATCH" in q.flags:
        base -= 0.10

    return float(max(0.05, min(0.95, base)))


def compute_alignment(q: EnrichedQuestion) -> str:
    # Alignment is only meaningful with matches
    if q.match_strength == "STRONG":
        return "HIGH"
    if q.match_strength == "WEAK":
        return "MED"
    return "LOW"
```

```python
def generate_answers_node(state: TenderState) -> TenderState:
    enriched_out: List[EnrichedQuestion] = []

    for q in state.enriched:
        try:
            # Start flags with routing information
            flags = list(q.flags)

            # Set match-based flags (useful later)
            if q.match_strength == "NONE":
                if "NO_STRONG_HISTORICAL_MATCH" not in flags:
                    flags.append("NO_STRONG_HISTORICAL_MATCH")

            # Branch prompt
            if q.route == "HISTORICAL_GUIDED":
                prompt = build_historical_guided_prompt(q)
            else:
                prompt = build_template_safe_prompt(q)

            # LLM draft
            draft = gen_llm.invoke(prompt).content.strip()

            # Apply deterministic guardrails
            safe_answer, extra_flags = apply_guardrails(q.question, draft, q.risk_hints)
            flags = sorted(set(flags) | set(extra_flags))

            # If guardrails were triggered, we also nudge alignment down slightly
            alignment = compute_alignment(q)
            if "HIGH_RISK" in flags and alignment == "HIGH":
                alignment = "MED"

            # Store answer + confidence
            q.answer = safe_answer
            q.flags = flags
            q.alignment = alignment
            q.confidence = compute_confidence(q)

            enriched_out.append(q)

        except Exception as e:
            q.error = str(e)
            q.flags = sorted(set(q.flags) | {"GENERATION_ERROR"})
            q.answer = ""
            q.confidence = 0.0
            q.alignment = "LOW"
            enriched_out.append(q)

    state.enriched = enriched_out
    return state
```

```python
from langgraph.graph import StateGraph, END
import uuid

graph = StateGraph(TenderState)
graph.add_node("classify_retrieve_route", classify_retrieve_route_node)
graph.add_node("generate_answers", generate_answers_node)
graph.add_node("summarise", summarise_node)

graph.set_entry_point("classify_retrieve_route")
graph.add_edge("classify_retrieve_route", "generate_answers")
graph.add_edge("generate_answers", "summarise")
graph.add_edge("summarise", END)

app_graph = graph.compile()
```

```python
class TenderState(BaseModel):
    request_id: str
    questions: List[QuestionItem] = Field(default_factory=list)

    # Store as dicts to avoid nested model validation issues in LangGraph
    enriched: List[dict] = Field(default_factory=list)

    summary: RunSummary = Field(default_factory=RunSummary)

    retrieval_top_k: int = 3
    strong_threshold: float = 0.70
    weak_threshold: float = 0.60
```

```python
def classify_retrieve_route_node(state: TenderState) -> TenderState:
    out: List[dict] = []

    for item in state.questions:
        try:
            dec = hybrid_classify(item.question)

            hits = vectordb.similarity_search_with_relevance_scores(
                item.question, k=state.retrieval_top_k
            )

            matches: List[dict] = []
            for doc, score in hits:
                md = doc.metadata or {}
                matches.append({
                    "source_id": str(md.get("source_id", "")),
                    "domain": str(md.get("domain", "")),
                    "historical_question": str(md.get("historical_question", "")),
                    "historical_answer": str(md.get("historical_answer", "")),
                    "similarity": float(score),
                })

            top_sim = matches[0]["similarity"] if matches else 0.0
            strength = match_strength(top_sim, strong=state.strong_threshold, weak=state.weak_threshold)
            route = route_from_strength(strength)

            out.append({
                "question_id": item.question_id,
                "question": item.question,
                "domain": dec.domain,
                "domain_confidence": float(dec.confidence),
                "domain_used": dec.used,
                "domain_rationale": dec.rationale,
                "risk_hints": dec.risk_hints,
                "matches": matches,
                "top_similarity": float(top_sim),
                "match_strength": strength,
                "route": route,

                # generation fields (initialized)
                "answer": "",
                "confidence": 0.0,
                "alignment": "LOW",
                "flags": [],
                "error": None,
            })

        except Exception as e:
            out.append({
                "question_id": item.question_id,
                "question": item.question,
                "domain": item.suggested_domain or "General",
                "domain_confidence": 0.0,
                "domain_used": "rules",
                "domain_rationale": "Failed during classification/retrieval.",
                "risk_hints": [],
                "matches": [],
                "top_similarity": 0.0,
                "match_strength": "NONE",
                "route": "TEMPLATE_SAFE",
                "answer": "",
                "confidence": 0.0,
                "alignment": "LOW",
                "flags": ["PIPELINE_ERROR"],
                "error": str(e),
            })

    state.enriched = out
    return state
```

```python
def generate_answers_node(state: TenderState) -> TenderState:
    out: List[dict] = []

    for q in state.enriched:
        try:
            flags = list(q.get("flags", []))
            risk_hints = q.get("risk_hints", [])
            route = q.get("route", "TEMPLATE_SAFE")
            domain = q.get("domain", "General")
            match_strength_val = q.get("match_strength", "NONE")

            # Ensure no-match flag
            if match_strength_val == "NONE":
                if "NO_STRONG_HISTORICAL_MATCH" not in flags:
                    flags.append("NO_STRONG_HISTORICAL_MATCH")

            # Build prompt
            if route == "HISTORICAL_GUIDED":
                # build evidence from matches
                evidence_blocks = []
                for m in (q.get("matches") or [])[:3]:
                    evidence_blocks.append(
                        f"- Source: {m.get('source_id')} | Domain: {m.get('domain')} | Similarity: {m.get('similarity'):.3f}\n"
                        f"  Historical Q: {m.get('historical_question')}\n"
                        f"  Historical A: {m.get('historical_answer')}\n"
                    )
                evidence = "\n".join(evidence_blocks)

                prompt = f"""
You are generating a tender response. Use the historical answers as guidance and remain consistent with them.
Do NOT invent certifications, certificate numbers, audit results, uptime SLAs, or legal compliance claims unless explicitly supported by the evidence.
Keep the response professional, concise, and aligned to the question.

Question:
{q['question']}

Detected Domain: {domain}
Risk Hints: {risk_hints}

Historical Evidence:
{evidence}

Write a response that answers the question directly. If the evidence is insufficient for a claim, state that details can be provided during due diligence or subject to contract.
""".strip()
            else:
                template = DOMAIN_TEMPLATES.get(domain, DOMAIN_TEMPLATES["General"])
                prompt = f"""
You are generating a tender response where there is no strong historical match.
Use the provided safe template as a baseline, adapt it to the question, and keep language cautious.
Do NOT invent certifications, certificate numbers, audit results, uptime SLAs, or legal compliance claims.

Question:
{q['question']}

Detected Domain: {domain}
Risk Hints: {risk_hints}

Safe Baseline Template:
{template}

Write a professional response. If specifics are required, state they can be provided during due diligence or agreed contractually.
""".strip()

            draft = gen_llm.invoke(prompt).content.strip()

            safe_answer, extra_flags = apply_guardrails(q["question"], draft, risk_hints)
            flags = sorted(set(flags) | set(extra_flags))

            # alignment based on match strength
            if match_strength_val == "STRONG":
                alignment = "HIGH"
            elif match_strength_val == "WEAK":
                alignment = "MED"
            else:
                alignment = "LOW"

            if "HIGH_RISK" in flags and alignment == "HIGH":
                alignment = "MED"

            # confidence
            # compute similarly, but on dict
            base = 0.45
            base += 0.25 * float(q.get("domain_confidence", 0.0))
            if match_strength_val == "STRONG":
                base += 0.25
            elif match_strength_val == "WEAK":
                base += 0.15
            else:
                base += 0.05
            if "HIGH_RISK" in flags:
                base -= 0.20
            if "NEEDS_EVIDENCE" in flags:
                base -= 0.10
            if "NO_STRONG_HISTORICAL_MATCH" in flags:
                base -= 0.10
            confidence = float(max(0.05, min(0.95, base)))

            # write back
            q["answer"] = safe_answer
            q["flags"] = flags
            q["alignment"] = alignment
            q["confidence"] = confidence
            q["error"] = None

            out.append(q)

        except Exception as e:
            q["error"] = str(e)
            q["flags"] = sorted(set(q.get("flags", [])) | {"GENERATION_ERROR"})
            q["answer"] = ""
            q["confidence"] = 0.0
            q["alignment"] = "LOW"
            out.append(q)

    state.enriched = out
    return state
```

```python
def summarise_node(state: TenderState) -> TenderState:
    state.summary.total = len(state.questions)
    state.summary.errors = sum(1 for x in state.enriched if x.get("error"))

    state.summary.routed_historical = sum(1 for x in state.enriched if x.get("route") == "HISTORICAL_GUIDED")
    state.summary.routed_template = sum(1 for x in state.enriched if x.get("route") == "TEMPLATE_SAFE")
    state.summary.flagged_risk = sum(1 for x in state.enriched if x.get("risk_hints"))

    state.summary.status = "PARTIAL" if state.summary.errors > 0 else "COMPLETED"
    return state
```
